```python
import math
raio = int(input("Digite um número interiro: "))
area = math.pi*raio**2
if raio > 0:
 print("A área do círculo de raio {} é {:.2f}". format(raio,area))
else:
   print("\033[91mPROBLEMA: não existe area do circulo para valor de raio negativo\033[0m")
```
