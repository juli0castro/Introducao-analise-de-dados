```python
import math
angulo = float(input("Digita o valor do angulo em graus:"))
angulorad =(angulo*math.pi)/180
seno= math.sin(angulorad)
cosseno = math.cos(angulorad)
tangente = math.tan(angulorad)
print ("Para esse angulo, temos o seno igual a {}, cosseno igual a {} e tangente igual a {}". format(seno,cosseno,tangente))
```

```python
import math
aplicmensal = float(input("Digite o valor constante da aplicação mensal: "))
taxanumber = float(input("Qual a taxa de rendimento: "))
taxa = taxanumber/100
nmeses= float(input("Qual o numero de meses:"))
rendimento = aplicmensal*(((1+taxa)**(nmeses)-1)/taxa)
print ("Seu rendimento será de {:.2f} reais".format(rendimento))
```

```python
numseg = int(input("Digite o numero de segundos:"))
horas = numseg//3600
resthoras = numseg%3600
minutos = resthoras//60
restminutos = resthoras%60
segundos = restminutos
print("Esse numero de segundos representa {:.0f}h{:.0f}min{}s".format(horas,minutos,segundos))
```

```python
import math
A = int(input("Digite o  coeficiete A: "))
B = int(input("Digite o coeficiete B: "))
C= int(input("Digite o coeficiete C: "))
discrim = ((B**2)-4*A*C)
if discrim < 0:
 print ("A equação não possui raízes reais.")
else:
 raiz1 = (-B+(discrim)**0.5)/(2*A)
 raiz2 = (-B-(discrim)**(0.5))/2*A
 print("A raiz 1 da equação de segundo grau é {} e a raiz 2 é {}". format (raiz1,raiz2))

```

```python
ladoA = float(input("Digite o comprimento do lado A do triângulo: "))
ladoB = float(input("Digite o comprimento do lado B do triângulo: "))
ladoC = float(input("Digite o comprimento do lado C do triângulo: "))
if (ladoA == ladoB == ladoC):
  print ("o triangulo é equilatero")
elif ladoA == ladoB or ladoA == ladoB or ladoA == ladoC or ladoB == ladoC:
    print ("o triangulo é isóceles")
else:
  print ("o triangulo é escaleno")
```
